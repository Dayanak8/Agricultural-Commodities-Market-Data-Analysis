```python
import pandas as pd


df['Price Date'] = pd.to_datetime(df['Price Date'])


state_season_categories = {
    'Tamil Nadu': {1: 'Winter', 2: 'Winter', 3: 'Summer', 
                   4: 'Summer', 5: 'Summer', 6: 'Mansoon', 
                   7: 'Mansoon', 8: 'Mansoon', 9: 'Mansoon', 
                   10: 'Mansoon', 11: 'Mansoon', 12: 'Mansoon'},
    'Karnataka': {1: 'Winter', 2: 'Winter', 3: 'Summer', 
                  4: 'Summer', 5: 'Summer', 6: 'Mansoon', 
                  7: 'Mansoon', 8: 'Mansoon', 9: 'Mansoon', 
                  10: 'Mansoon', 11: 'Mansoon', 12: 'Mansoon'},
    'Kerala': {1: 'Winter', 2: 'Winter', 3: 'Summer', 
               4: 'Summer', 5: 'Summer', 6: 'Mansoon', 
               7: 'Mansoon', 8: 'Mansoon', 9: 'Mansoon', 
               10: 'Mansoon', 11: 'Mansoon', 12: 'Winter'},
    
}

# Function to assign seasonality indicator based on state and month
def get_seasonality_indicator(row):
    state = row['State Name']
    month = row['Price Date'].month
    return state_season_categories[state].get(month, 'Unknown')

df['Season'] = df.apply(get_seasonality_indicator, axis=1)

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

grouped_data = df.groupby(['State Name', 'Commodity']).size().reset_index(name='Count')

# Get top 20 commodities for each state
states = df['State Name'].unique()
top_commodities_per_state = {}
for state in states:
    top_commodities_per_state[state] = grouped_data[grouped_data['State Name'] == state].nlargest(20, 'Count')

for state, top_commodities in top_commodities_per_state.items():
    plt.figure(figsize=(10, 10))
    sns.barplot(x='Count', y='Commodity', data=top_commodities)
    plt.xlabel("Count", fontsize=14)
    plt.ylabel("Commodity", fontsize=14)
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    plt.title(f"Top 20 Commodities in {state}", fontsize=16)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

categories = ['Planation crops', 'Grains', 'Cereals and Pulses', 'Fruits', 'Vegetables',
              'Spices', 'Nuts and Oilseeds', 'Others', 'Livestock and Animal Products', 'Flower']
years = [2020, 2021, 2022, 2023]
df_filtered = df[df['Category'].isin(categories) & df['Price Date'].dt.year.isin(years)]


grouped_data = df_filtered.groupby(['State Name', 'Category', df_filtered['Price Date'].dt.year])['modal_price'].mean().reset_index()

# Pivot the DataFrame to have years as columns
pivot_data = grouped_data.pivot_table(index=['State Name', 'Category'], columns='Price Date', values='modal_price')

# Plot modal price variation over the years for each category and state
for state, state_data in pivot_data.groupby(level=0):
    plt.figure(figsize=(12, 8))
    state_data.plot(kind='bar', ax=plt.gca())
    plt.title(f'Modal Price Variation Over Years (2020-2023) for Each Category in {state}')
    plt.xlabel('Category')
    plt.ylabel('Average Modal Price')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filtered_data = df[df['Category'].isin(['Planation crops', 'Spices'])]

if not filtered_data.empty:

    grouped_data = filtered_data.groupby(['State Name', 'Category', 'Commodity', 'Variety'])['modal_price'].mean().reset_index()

    result = grouped_data.loc[grouped_data.groupby(['State Name', 'Category', 'Commodity'])['modal_price'].idxmax()]
    

    for state in df['State Name'].unique():
        state_data = result[result['State Name'] == state]
        
        # Plotting
        plt.figure(figsize=(12, 8))
        for category, category_data in state_data.groupby('Category'):
            max_price_variety = category_data.groupby('Commodity')['modal_price'].idxmax()
            max_price_data = category_data.loc[max_price_variety]
            plt.bar(max_price_data['Commodity'] + ' (' + max_price_data['Variety'] + ')', max_price_data['modal_price'], label=category)
        plt.xlabel('Commodity with Highest Modal Price (Variety)')
        plt.ylabel('Average Modal Price')
        plt.title(f'Highest Modal Price Commodity by Category in {state}')
        plt.xticks(rotation=45, ha='right')
        plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
else:
    print("No data available for the specified categories")
```
