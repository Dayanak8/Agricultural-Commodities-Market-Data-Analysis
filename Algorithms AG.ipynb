# AR

```python
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


df = pd.read_csv("model.csv") 

# Convert 'Year' column to integer if it's not already in integer format
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Function to generate forecast for all categories
def generate_forecast():
    category_data_yearly_mean = df.groupby(['Category', 'Year'])['modal_price'].mean().reset_index()
    
    categories = category_data_yearly_mean['Category'].unique()
    
    rmse_list = []
    r_score_list = []
    
    for category in categories:
       
        category_data = category_data_yearly_mean[category_data_yearly_mean['Category'] == category]
        
        try:
            # Fit AR model
            model = AutoReg(category_data['modal_price'], lags=1) 
            ar_results = model.fit()
            
            # Forecast data for future years
            forecast_steps = len(category_data)  # Forecast for the length of available data
            forecast = ar_results.predict(start=1, end=forecast_steps)
            
            # Calculate Mean Squared Error (MSE)
            mse = mean_squared_error(category_data['modal_price'], forecast)
            
            # Calculate Root Mean Squared Error (RMSE)
            rmse = np.sqrt(mse)
            
            # Calculate R-squared (r_score)
            r_score = np.abs(r2_score(category_data['modal_price'], forecast))  
            
            rmse_list.append(rmse)
            r_score_list.append(r_score)

        except Exception as e:
          
            print(f"An error occurred for {category}: {e}")

   
    overall_rmse = np.mean(rmse_list)
    
 
    overall_r_score = np.mean(r_score_list)

    print(f"AR RMSE: {overall_rmse}")
    print(f"AR R-squared (r_score): {overall_r_score}")


generate_forecast()
```

# ARIMA

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')


def generate_forecast():
   
    category_data_yearly_mean = df.groupby(['Category', 'Year'])['modal_price'].mean().reset_index()
    
    categories = category_data_yearly_mean['Category'].unique()
    
    rmse_list = []
    r_score_list = []
    
    for category in categories:

        category_data = category_data_yearly_mean[category_data_yearly_mean['Category'] == category]
        
        try:
            # Fit ARIMA model
            model = ARIMA(category_data['modal_price'], order=(1, 0, 0))  # Use ARIMA(1, 0, 0) for AR model
            arima_results = model.fit()

            # Forecast data for future years
            forecast_steps = len(category_data)  
            forecast = arima_results.predict(start=1, end=forecast_steps)
            
            # Calculate Mean Squared Error (MSE)
            mse = mean_squared_error(category_data['modal_price'], forecast)
            
            # Calculate Root Mean Squared Error (RMSE)
            rmse = np.sqrt(mse)

            # Calculate R-squared (r_score)
            r_score = np.abs(r2_score(category_data['modal_price'], forecast))  

            rmse_list.append(rmse)
            r_score_list.append(r_score)

        except Exception as e:
        
            print(f"An error occurred for {category}: {e}")

   
    overall_rmse = np.mean(rmse_list)
    
   
    overall_r_score = np.mean(r_score_list)

    print(f"ARIMA RMSE: {overall_rmse}")
    print(f"ARIMA R-squared (r_score): {overall_r_score}")

generate_forecast()
```

# SARIMA

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

df = pd.read_csv("model.csv") 

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

def generate_forecast_sarima():
    category_data_yearly_mean = df.groupby(['Category', 'Year'])['modal_price'].mean().reset_index()
    
    categories = category_data_yearly_mean['Category'].unique()
    
    rmse_list = []
    r_score_list = []
    
    for category in categories:
       
        category_data = category_data_yearly_mean[category_data_yearly_mean['Category'] == category]
        
        try:
            # Fit SARIMA model
            model = SARIMAX(category_data['modal_price'], order=(1, 0, 0), seasonal_order=(1, 0, 0, 12))  # SARIMA(1, 0, 0)(1, 0, 0, 12)
            sarima_results = model.fit()

            # Forecast data for future years
            forecast_steps = len(category_data)  
            forecast = sarima_results.predict(start=1, end=forecast_steps)
            
            # Calculate Mean Squared Error (MSE)
            mse = mean_squared_error(category_data['modal_price'], forecast)
            
            # Calculate Root Mean Squared Error (RMSE)
            rmse = np.sqrt(mse)

            # Calculate R-squared (r_score)
            r_score = np.abs(r2_score(category_data['modal_price'], forecast))  

            rmse_list.append(rmse)
            r_score_list.append(r_score)

        except Exception as e:
      
            print(f"An error occurred for {category}: {e}")

    # Calculate overall RMSE
    overall_rmse = np.mean(rmse_list)
    
    # Calculate overall R-squared (r_score)
    overall_r_score = np.mean(r_score_list)

    print(f"SARIMA RMSE: {overall_rmse}")
    print(f"SARIMA R-squared (r_score): {overall_r_score}")

generate_forecast_sarima()
```

```python

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

def generate_forecast():
 
    category_data_yearly_mean = df.groupby(['Category', 'Year'])['modal_price'].mean().reset_index()
    
    categories = category_data_yearly_mean['Category'].unique()
    
    forecasts_df = pd.DataFrame(columns=['Category', 'Year', 'Forecast'])
    
    for category in categories:
        category_data = category_data_yearly_mean[category_data_yearly_mean['Category'] == category]
        
        try:
            # Fit SARIMA model
            model = SARIMAX(category_data['modal_price'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)) 
            sarima_results = model.fit()

            # Forecast data for future years
            forecast_steps = 1  # Forecast for the next 1 year
            forecast = sarima_results.forecast(steps=forecast_steps)
            
            # Add the forecast to the DataFrame
            forecast_df = pd.DataFrame({'Category': [category]*forecast_steps,
                                        'Year': range(category_data['Year'].max()+1, category_data['Year'].max()+1+forecast_steps),
                                        'Forecast': forecast})
            forecasts_df = pd.concat([forecasts_df, forecast_df], ignore_index=True)

        except Exception as e:
            # Print error message
            print(f"An error occurred for {category}: {e}")

    return forecasts_df

# Example usage
forecasts_df = generate_forecast()
print(forecasts_df)
```

# GRADIENT BOOSTING REGRESSOR

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


df = pd.read_csv("model.csv") 

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

X = df[['Year', 'Category']]
y = df['modal_price']

X_encoded = pd.get_dummies(X, columns=['Category'], drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=20)

gb_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_regressor.fit(X_train, y_train)

# Make predictions
y_pred = gb_regressor.predict(X_test)

# Calculate Mean Squared Error and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

# CATBOOST REGRESSOR

```python
import pandas as pd
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("model.csv")  


df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

X = df[['Year', 'Category']]
y = df['modal_price']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

# Initialize and train the CatBoostRegressor model
catboost_regressor = CatBoostRegressor(iterations=1000, learning_rate=0.1, depth=6, loss_function='RMSE')
catboost_regressor.fit(X_train, y_train, cat_features=['Category'], verbose=False)

# Make predictions
y_pred = catboost_regressor.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```
